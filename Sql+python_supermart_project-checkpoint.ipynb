# 2. Total sales for each branch

```python
query_2 = "SELECT Branch, SUM(Total) AS TotalSales FROM supermarket_sales1 GROUP BY Branch ORDER BY TotalSales DESC"
cur.execute(query_2)
d1=cur.fetchall()
df= pd.DataFrame(d1, columns=["Branch","TotalSales"])
plt.figure(figsize=(4, 4))
plt.pie(df["TotalSales"], labels=df["Branch"], autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=140)

# Title and display
plt.title("Total Sales by Branch")
plt.show()
```

# 6. Average unit price of each product category

```python
query6 =  "SELECT Product_Line, AVG(Unit_Price) AS AvgPrice FROM supermarket_sales1 GROUP BY Product_Line;"
cur.execute(query6)
d1=cur.fetchall()
df= pd.DataFrame(d1, columns=["Product_Line","AvgPrice"])
sns.set(style="whitegrid")

# Create the bar chart
plt.figure(figsize=(10, 4))
sns.barplot(y=df["Product_Line"], x=df["AvgPrice"], hue=df["AvgPrice"],palette="coolwarm", edgecolor="black")

# Labels and title
plt.xlabel("Average Unit Price")
plt.ylabel("Product Line")
plt.title("Average Unit Price by Product Line")

# Show chart
plt.show()
```

# 19. Calculate the average rating per product line
